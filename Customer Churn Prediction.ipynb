```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
Cust_data = pd.read_csv('Churn_Modelling.csv')
```

```python
Cust_data.head()
```

```python
Gender_mapping = {'Male': 1,'Female': 2}
Cust_data['Gender'] = Cust_data['Gender'].map(Gender_mapping)
```

```python
Cust_data.tail()
```

```python
Geography_mapping = {'France': 1,'Spain': 2, 'Germany': 3}
Cust_data['Geography'] = Cust_data['Geography'].map(Geography_mapping)
```

```python
Cust_data.head()
```

```python
Cust_data = Cust_data.drop(['RowNumber','Surname','CustomerId'], axis=1)
```

```python
Cust_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = Cust_data.drop('Exited', axis=1)
Y = Cust_data['Exited']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
Model = LogisticRegression(max_iter=1000, random_state=42)
```

```python
Model.fit(X_train, Y_train)
```

```python
Y_pred = Model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test,Y_pred)
print(f"Accuracy Score:, {accuracy}")
```

# Accuracy Score: 80%
